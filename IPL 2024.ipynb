```python
import pandas as pd

# Assuming ipl_data is your DataFrame
ipl_data['Year'] = pd.to_datetime(ipl_data['Year'], format='%Y', errors='coerce')

# Now check for any NaT values
print(ipl_data[ipl_data['Year'].isna()])
```

```python
# Handling missing values
# Since the dataset is about player performance, missing values in numerical columns can be filled with 0
numerical_cols = ['Matches_Batted', 'Not_Outs', 'Runs_Scored', 'Highest_Score', 'Batting_Average',
                  'Balls_Faced', 'Batting_Strike_Rate', 'Centuries', 'Half_Centuries', 'Fours', 'Sixes',
                  'Catches_Taken', 'Stumpings', 'Matches_Bowled', 'Balls_Bowled', 'Runs_Conceded',
                  'Wickets_Taken', 'Best_Bowling_Match', 'Bowling_Average', 'Economy_Rate',
                  'Bowling_Strike_Rate', 'Four_Wicket_Hauls', 'Five_Wicket_Hauls']
ipl_data[numerical_cols] = ipl_data[numerical_cols].fillna(0)
```

```python
# Filter out non-numeric values from 'Runs_Scored' column
numeric_runs = pd.to_numeric(ipl_data['Runs_Scored'], errors='coerce').dropna()

plt.figure(figsize=(12, 8))

sns.histplot(numeric_runs, bins=30, kde=True, color='skyblue', edgecolor='black', linewidth=1.2)

# Add mean, median, and standard deviation
mean_runs = np.mean(numeric_runs)
median_runs = np.median(numeric_runs)
std_dev = np.std(numeric_runs)

plt.axvline(mean_runs, color='red', linestyle='--', label=f'Mean: {mean_runs:.2f}')
plt.axvline(median_runs, color='green', linestyle='--', label=f'Median: {median_runs}')
plt.axvline(mean_runs + std_dev, color='purple', linestyle='--', label=f'Std Dev: {std_dev:.2f}')
plt.axvline(mean_runs - std_dev, color='purple', linestyle='--')

# Highlight outliers
outliers = numeric_runs[numeric_runs > mean_runs + 2 * std_dev]
plt.scatter(outliers, np.zeros_like(outliers), color='red', label='Outliers')
plt.title('Distribution of Runs Scored by Players')
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Convert columns to numeric
ipl_data['Batting_Average'] = pd.to_numeric(ipl_data['Batting_Average'], errors='coerce')
ipl_data['Runs_Scored'] = pd.to_numeric(ipl_data['Runs_Scored'], errors='coerce')

# Remove rows with missing or non-numeric values
ipl_data = ipl_data.dropna(subset=['Batting_Average', 'Runs_Scored'])

plt.figure(figsize=(10, 6))
sns.regplot(x='Batting_Average', y='Runs_Scored', data=ipl_data,
            scatter_kws={'s': 100, 'alpha': 0.5, 'color': 'blue'}, line_kws={'color': 'red'})
plt.title('Relationship between Batting Average and Runs Scored')
plt.xlabel('Batting Average')
plt.ylabel('Runs Scored')
plt.legend(['Regression Line'])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'data' containing player names and their runs scored
# Replace 'data' with the actual name of your DataFrame
top_10_runs = ipl_data.sort_values(by='Runs_Scored', ascending=False).head(10)

plt.figure(figsize=(20, 10))
sns.set_style("whitegrid")

# Specify a single color for the bars
sns.barplot(x='Runs_Scored', y='Player_Name', data=top_10_runs, color='purple', alpha=0.8)

for i, (value, name) in enumerate(zip(top_10_runs['Runs_Scored'], top_10_runs['Player_Name'])):
    plt.text(value + 5, i, f'{value}', ha='left', va='center', fontsize=12, color='black', fontweight='bold')

plt.show()
```
